### 手書き文字データのインポート、ライブラリのインポート

```

from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
```

### 手書き文字画像データの変数化

```
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
```

```
# フレームワークの提供する高速な処理を実行できるC++を利用する際のオブジェクト
sess = tf.InteractiveSession()
```

### 入力情報を持たせるためにplaceholderを定義する。
### shape(第2引数)の次元にNoneを指定すると、どんな長さの次元数であっても対応できる。

```
x = tf.placeholder(tf.float32, [None, 784])
```

### 重みとバイアスを保持するVariableを定義する。
* Variable関数は値を修正する。
* それぞれ初期値は今回0で。

```
# 重み 784次元の入力を受けて10次元の出力を返す
W = tf.Variable(tf.zeros([784, 10]))

# バイアス 10次元を用意する
b = tf.Variable(tf.zeros([10])) 
```

### ニューラルネットモデルを定義する。
### 入力xと重みWの行列積(tf.matmul)の出力にバイアスbを加え、ソフトマックスで最終的な出力を決定する。

```
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

### 以上でニューラルネットモデルの定義は終わり。
### 以下は、訓練・評価の実装
### (今回のサンプルでは、交差エントロピーで評価して学習しています。)

```
# 教師データ(正しい答え)を保持するplaceholderを定義する。
y_ = tf.placeholder(tf.float32, [None, 10])
```

### 正解との差分をcross_entropyとして、

```
# 交差エントロピーの計算式を定義する。
# 教師データy_とモデルからの出力yの対数をとったものとの積を取り、全体の合計を計算している
cross_entropy = -tf.reduce_sum(y_ * tf.log(y))
```

### 各ステップにおけるモデルの更新方法を定義してます。
* 1ステップごとに0.01の学習率で交差エントロピーが最小になるようにする
* y = tf.nn.softmax(...)は確率をもとめてます


```
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```
# すべての変数を初期化するための準備をする。
tf.initialize_all_variables().run()
```

```
# Sessionを定義し、すべての変数を初期化する。
sess = tf.Session()
sess.run(init)
```

# 訓練をしてます↓

```

for i in range(1000):
  batch_xs, batch_ys = mnist.train.next_batch(100)
  train_step.run({x: batch_xs, y_: batch_ys})
```

### 予測が正解と合っているか確認するためにtf.equal を使ってます

```
# 性能評価のための評価式を定義する。
# 入力データに対するモデルの出力yと教師データy_が一致しているか確認する。
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
```

```
# 最終的な性能評価は平均値で決定することを定義する。
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
```

```
print(accuracy.eval({x: mnist.test.images, y_: mnist.test.labels}))
```
